### Trigonometric Functions
---

#### Basic Use of Trig Functions

First, note that the trigonometric functions can be found in the **numpy** library.  We will use the syntax `np.sin()` or `np.cos()` to find the value of a trigonometric function.  

```python
np.sin(0), np.cos(3.14)
```

Angles are given in radians, so it is important to remember how to convert back and forth.  

$$\text{radians} = \frac{\text{degrees}\times \pi}{180}$$

If we want to find the sine of 90 degrees, we convert this to radians and enter it into the sin function.  Note that $\pi$ is also a numpy function.  

An easy way to do this is to assign a variable the angle value so you can substitute this into the sine function.

```python
a = 90*np.pi/180
```

```python
np.sin(a)
```

#### Plotting Trig Functions

We use our familiar matplotlib plotting library and approach.  Noting that these accept radian inputs, we can plot a single period of each function by defining our inputs as `t = np.linspace(0,2*np.pi,1000)`.

```python
t = np.linspace(0,2*np.pi, 1000)
plt.plot(t,np.sin(t), label="$sin(t)$")
plt.plot(t,np.cos(t), label="$cos(t)$")
plt.title("Sine and Cosine on $[0,2\pi]$") #adds title
plt.legend(loc='best') #places legend
```

#### Key Features I
---

**Amplitude**: Half the distance between the highest and lowest points of the functions.  

**Frequency**: How many cycles in one unit of time.

**Period**: How much time for one cycle.

We have:

$$y = A\sin{2\pi tf}$$

where $A$ is the amplitude and $f$ is the frequency.  For example, the graph of $y = 10\sin{\pi t}$ has amplitude 10 and frequency $\frac{1}{2}$.

```python
def f(t):
    return 10*np.sin(np.pi*t)
```

```python
t = np.linspace(0,np.pi,1000)
```

```python
plt.plot(t,f(t))
```

#### Phase Shift
---

A phase shift is a way to consider a periodic situation that starts somewhere other than an angle of zero.  Perhaps, for example, a Ferris Wheel is being modeled and we start at the top.  This means we are 90 degrees or $\pi/2$ out of phase.  We incorporate this into our model with a horizontal translation, i.e.: 

$$y = a\sin({2\pi ft + phase})$$

```python
def g(t):
    return 10*np.sin(np.pi/4 * t)

def g2(t):
    return 10*np.sin(np.pi/4 * t + np.pi/4)

def g3(t):
    return 10*np.sin(np.pi/4 * t + np.pi/2)
    
```

```python
plt.plot(t,g(t))
plt.plot(t,g2(t))
plt.plot(t,g3(t))
```
